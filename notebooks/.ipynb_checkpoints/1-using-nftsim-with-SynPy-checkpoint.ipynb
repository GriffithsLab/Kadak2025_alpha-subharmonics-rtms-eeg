```python
%%time
import os
os.environ['OMP_NUM_THREADS'] = '1'
import SynPy as sp
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.patches as patches
# import matplotlib.cm as cm
import re
from tqdm import tqdm
import seaborn as sns
import multiprocessing as mp
import fnmatch
import pickle
from sklearn import preprocessing
from scipy.stats import linregress

#--Run Options--#
conf_dir = os.path.join(os.getcwd(), 'confs/')
output_dir = os.path.join(os.getcwd(), 'outputs/')


grid_dir_name = 'bursts_oscillation_fixed'
new_conf_dir = os.path.join(conf_dir, grid_dir_name)
new_output_dir = os.path.join(output_dir, grid_dir_name)
```

# Working with NFTsim revolves around .conf (configuation) files, which contain the parameter and hyperparameters for simulation

```python
# First, lets store a .conf file as a SynPy dot_conf class object
conf_obj = sp.dot_conf('eirs-tms-custom.conf')

# We can do a few things with this. Including:

# File path
print(conf_obj.conf_path)
print()

# File name
print(conf_obj.f_name)
print()

# File contents
conf_obj.conf_text
```

```python
# Next, lets run the conf with nftsim and store its output, which will be a dot_output class object.  We can also directly start working with .output files, which will be useful later
output_obj = conf_obj.run()

# Here, the object is storing the time series of each population (pop._.q/v) and connection (coupling._.nu/nutilde/gnmda, etc.) variable in the simulation
print(output_obj)

# Lets look at a timeseries of the voltage of the excitatory cortical population, which will spike strongy during the active period of TMS
plt.figure(figsize=(10, 4))
plt.plot(output_obj['pop.e.v'], color='mediumseagreen')
plt.title('Membrane Voltage of Excitatory Cortical Population (E)')
plt.xlabel('Time (ms)')
plt.ylabel('Voltage (mV)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Let's also look at the dynamics of a connection between two populations, such as ei (incoming-outgoing; post-presynaptic)

# First plot: Intracellular calcium dynamics
plt.figure(figsize=(10, 4))
plt.plot(output_obj['coupling.ei.ca'], color='dodgerblue')
plt.title('Intracellular Calcium in ei Connection')
plt.xlabel('Time (ms)')
plt.ylabel('Calcium Concentration')
plt.grid(True)
plt.tight_layout()
plt.show()


# This produces plasticity effects, which we can observe in the pre (nutilde) and post (nu) kinase cascade signals, representing physiologically unrealized and realized synaptic weight change following a delay
# Second plot: Synaptic weight (nu, nutilde) dynamics
plt.figure(figsize=(10, 4))
plt.plot(output_obj['coupling.ei.nu'], color='purple')
plt.plot(output_obj['coupling.ei.nutilde'], color='hotpink')
plt.title('Synaptic Coupling Strength (nu) in ei Connection')
plt.xlabel('Time (ms)')
plt.ylabel('Coupling Strength (nu)')
plt.grid(True)
plt.tight_layout()
plt.show()
# plt.plot(output_obj['coupling.ei.ca'])
```
